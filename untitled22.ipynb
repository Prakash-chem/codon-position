```python
# Import necessary libraries
import pandas as pd
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
import matplotlib.pyplot as plt
import seaborn as sns

# Function to read a FASTA file and return sequences with their headers
def read_fasta(file_path):
    names = []
    sequences = []
    with open(file_path, "r") as file:
        sequence = ""
        name = ""
        for line in file:
            line = line.strip()
            if line.startswith(">"):  # Header line
                if sequence:  # Save the previous sequence
                    sequences.append(sequence)
                    sequence = ""
                name = line[1:]  # Exclude '>'
                names.append(name)
            else:
                sequence += line
        if sequence:  # Add the last sequence
            sequences.append(sequence)
    return names, sequences

# Read your FASTA file
fasta_file = "107_whole.fasta"  # Replace with your actual FASTA file name
functional_names, sequences = read_fasta(fasta_file)

# Your new sequence (replace this with your actual new sequence)
new_sequence = "MEIRSYPFPDPDDLGVNETYLRLQREEPVSRVRLPYGGEAWIATRYADVKTVMGDPRFVRAPVMDPDRDPPRTMPEVPRISTILSMDPPEHTRLRRILAKAFTARRTAQLRDRAQEIVNDLLDRMEEQGPPADLMDALAMPVAITLISEMLGVPNLDRAQFRAWSDAAVAITAFTVPEIEAAAASLRAYIRDLVEEKRKHPADDMLTVLVTAHDDEDRLSEEELVSFGVTLLIAGHETTGSEIGNFMYTLLARPDKLAEVRADPSLLPTAIEELLRHTKLSGAPAFPRFATVDVELSGVRIAAGEAVFVDGLVANRDPEVFENPDELNFHRPSIPHFAFGHGVHHCIGAPLARMELQVAIATLLNRFPNLRLAGDVTIKKGRLVRGPQTLPVRW"

# Scoring parameters
match_score = 1
mismatch_score = -1
gap_open_score = -2
gap_extend_score = -0.5

# Function to calculate similarity and identity
def calculate_similarity_and_identity(seq1, seq2):
    alignments = pairwise2.align.globalms(seq1, seq2, match_score, mismatch_score, gap_open_score, gap_extend_score)
    best_alignment = alignments[0]

    # Similarity score: proportion of matched positions (includes gaps)
    similarity_score = best_alignment.score / max(len(seq1), len(seq2))

    # Identity score: proportion of identical matches (excludes gaps)
    matches = sum(res1 == res2 for res1, res2 in zip(best_alignment[0], best_alignment[1]) if res1 != '-' and res2 != '-')
    identity_score = matches / max(len(seq1), len(seq2))

    return similarity_score, identity_score

# Calculate similarity and identity for each sequence
scores = [calculate_similarity_and_identity(seq, new_sequence) for seq in sequences]
similarities, identities = zip(*scores)

# Create a DataFrame with results
results = pd.DataFrame({
    'Functional Name': functional_names,
    'Original Sequence': sequences,
    'Similarity Score': similarities,
    'Identity Score': identities
})

# Print the first few rows of the results
print(results)

# Optionally, save the results to a new Excel file
results.to_excel("alignment_results.xlsx", index=False)
print("Results saved to alignment_results.xlsx")
```

```python
# Sort by similarity score
results_sorted = results.sort_values('Similarity Score', ascending=False)

# Display top 10 results
print(results_sorted[['Functional Name', 'Similarity Score', 'Identity Score']].head(10))

# Create a heatmap of similarity scores
plt.figure(figsize=(12, 8))
sns.heatmap(results_sorted['Similarity Score'].values.reshape(1, -1),
            cmap='YlOrRd', annot=False, cbar=True)
plt.title('Similarity Scores Heatmap')
plt.xlabel('Sequence Index (sorted by similarity)')
plt.show()

# Create a bar plot of similarity scores
plt.figure(figsize=(12, 6))
plt.bar(range(len(similarities)), results_sorted['Similarity Score'])
plt.title('Similarity Scores for Each Sequence')
plt.xlabel('Sequence Index (sorted by similarity)')
plt.ylabel('Similarity Score')
plt.show()

# Create a bar plot of identity scores
plt.figure(figsize=(12, 6))
plt.bar(range(len(identities)), results_sorted['Identity Score'])
plt.title('Identity Scores for Each Sequence')
plt.xlabel('Sequence Index (sorted by identity)')
plt.ylabel('Identity Score')
plt.show()

# Save the sorted results
results_sorted.to_excel('sorted_similarity_and_identity_results.xlsx', index=False)

# Example of alignment visualization for the top sequence
top_sequence = results_sorted.iloc[0]['Original Sequence']
alignments = pairwise2.align.globalms(new_sequence, top_sequence, match_score, mismatch_score, gap_open_score, gap_extend_score)
for alignment in alignments:
    print(format_alignment(*alignment))
```
